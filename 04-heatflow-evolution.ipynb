```python
from heatlib_const import *
```

```python
m = dict(n=100, k=2.25, H=1e-6, rho=2700,
         c=800, tc=35000, T0=0, q=-0.02)
```

```python
init(m)
```

```python
m['t'][logical_and(m['x']>5000, m['x']<10000)] = 700
```

```python
hf = [1000*q_surface(m)]
time = [0]
dt = 1000*ysec
```

```python
for i in range(1000):
    for j in range(10):
        ftcs(m, dt)
    hf.append(1000*q_surface(m))
    time.append(m['time']/ma)
```

```python
plot(time, hf)
xlabel('Time [Ma]')
ylabel('Surface heatflow [mW/m2]')
```
