# Ukázka použití knihovny heatlib_var
Nejdřív importujeme knihovnu

```python
from heatlib_var import *
```

Pak vytvoříme slovník, který obsahuje informace o vlastnostech modelu.

  - n - počet uzlů v modelu
  - k - konduktivita (array n-1 elements)
  - H - producke tepla (array n-1 elements)
  - rho - hustota (array n-1 elements)
  - c - měrná tepelná kapacita (array n-1 elements)
  - tc - velikost domény (mocnost kůry)
  - T0 - teplota na povrchu (Dirichlet)
  - q - tepelný tok na bázi (Neumann)

```python
n = 100
k = 2.5*ones(n-1)
H = 1e-6*ones(n-1)
rho = 2700*ones(n-1)
c = 800*ones(n-1)
# definice slovniku pro model
m = dict(n=n, k=k, H=H, rho=rho, c=c, tc=35000, T0=0, q=-0.02)
```

Funkce `init` inicializuje model a vypočte stacionární řešení.

```python
init(m)
```

Funkce `tshow` zobrazí aktuální geotermu.

```python
tshow(m)
```

Modifikace stacionárního řešení. `t` a `x` jsou proměnné v modelu a obsahují aktuální teplotu a pozici uzlů

```python
m['t'][(m['x']<10000) & (m['x']>5000)] = 700
```

```python
tshow(m)
```

```python
dt = 1000*ysec
for j in range(100):
    btcs(m, dt)
```
